## <font color = red> Method 3 </font>


```python
sns.countplot(x = "country", data = data_set)
```

```python
sns.lineplot( data =data_set, x = "year" , y = "inflation_annual_cpi")
```

```python
sns.lineplot(data = data_set, x = "year" , y = "exch_usd")
```

```python
sns.swarmplot(x = "year" , y = "exch_usd" , data = data_set)
```

```python
sns.boxplot(x = "year" , y = "exch_usd" , data = data_set)
```

```python
data_set.groupby('year')['exch_usd'].sum().plot.bar(figsize=(20,10)).set(ylabel='exch_usd', xlabel='Year');
```

```python
data_set.plot.line(y='exch_usd', x='year',figsize=(18,7));
```

```python
data_set.exch_usd.hist()
plt.title('Histogram of exch_usd')
plt.xlabel('exch_usd')
plt.ylabel('Frequency')
```
